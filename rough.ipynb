```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel('CurrentStatus.xlsx')

# Extract the desired columns
epoch = df['epoch']
train_mae = df['Current Train MAE']
train_rmse = df['Current Train RMSE']
val_mae = df['Current Val MAE']
val_rmse = df['Current Val RMSE']

# Convert the DataFrame to a NumPy array
data = np.array([epoch, train_mae, train_rmse, val_mae, val_rmse])

# Create a plot
plt.figure(figsize=(8, 6))

# Plot the train MAE
plt.plot(epoch, train_mae, label='Train MAE')

# Plot the train RMSE
plt.plot(epoch, train_rmse, label='Train RMSE')

# Plot the val MAE
plt.plot(epoch, val_mae, label='Val MAE')

# Plot the val RMSE
plt.plot(epoch, val_rmse, label='Val RMSE')

# Set the title and labels
plt.title('Current Train and Val MAE/RMSE vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('MAE/RMSE')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
